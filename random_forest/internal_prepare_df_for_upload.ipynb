```python
df_to_write = df.loc[(~df["sample"].str.contains("old")) & (df["assign"].isin(["G", "C", "G/C", "U"]))]#.sort_values("sample_name").drop_duplicates("pseudo_prot")
```

```python
df_to_write["is_unsorted"] = df_to_write["sample_name"].str.contains("unsorted")
df_to_write = df_to_write[["pseudo_prot", "count", "num_muts", "lib", "freq", "enrichment", "sample_name", "is_unsorted", "assign"]]
df_to_write.to_csv("./NGS_data.csv")
df_to_write
```

```python
seq_space_df = pd.read_pickle(
    "/home/labs/fleishman/jonathaw/gfp_lib/analysis/deep-seq-analysis/nY_wY_NGS_210319/random_forest/seq_space_df_analysed.pkl"
).rename(columns={"pseudo_prot_temp": "pseudo_prot"})
seq_space_df["worked"] = seq_space_df["predicted"].progress_map(lambda x: x != "U")
seq_space_df["predicted_proba_func"] = 1 - seq_space_df["predicted_proba_U"]
seq_space_df.sort_values("predicted_proba_func", inplace=True, ascending=False)

seq_space_df["y"] = seq_space_df["predicted_proba_func"] > 0.5
for i, p in enumerate(all_poss):
    if p in seq_space_df.columns:
        seq_space_df[p] = seq_space_df["pseudo_prot"].str[i]

display_all(seq_space_df.head())
```

```python
merged_df= pd.merge(df, seq_space_df, on="pseudo_prot", how="outer")
display_all(merged_df)
```

```python
for i, p in enumerate(all_poss):
    merged_df[p] = merged_df["pseudo_prot"].str[i]
display_all(merged_df)
```
