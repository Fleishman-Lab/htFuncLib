```python
from pathlib import Path

import pandas as pd
from tqdm.notebook import tqdm
```

```python
def display_all(df):
    with pd.option_context("display.max_rows", 1000, "display.max_columns", 1000):
        display(df)
```

```python
all_poss = [
    "L42",
    "L44",
    "V61",
    "T65",
    "V68",
    "Q69",
    "S72",
    "Q94",
    "T108",
    "V112",
    "N121",
    "Y145",
    "H148",
    "V150",
    "T167",
    "H181",
    "N185",
    "T203",
    "S205",
    "L220",
    "E222",
    "V224",
]
ONE_2_THREE = {
    "A": "ALA",
    "C": "CYS",
    "D": "ASP",
    "E": "GLU",
    "F": "PHE",
    "G": "GLY",
    "H": "HIS",
    "I": "ILE",
    "K": "LYS",
    "L": "LEU",
    "M": "MET",
    "N": "ASN",
    "P": "PRO",
    "Q": "GLN",
    "R": "ARG",
    "S": "SER",
    "T": "THR",
    "V": "VAL",
    "W": "TRP",
    "Y": "TYR",
}
AA_CAT = {
    "A": "h_phobic",
    "C": "polar",
    "D": "negative",
    "E": "negative",
    "F": "aromatic",
    "G": "polar",
    "H": "aromatic",
    "I": "h_phobic",
    "K": "positive",
    "L": "h_phobic",
    "M": "h_phobic",
    "N": "polar",
    "P": "polar",
    "Q": "polar",
    "R": "positive",
    "S": "polar",
    "T": "polar",
    "V": "h_phobic",
    "W": "aromatic",
    "Y": "aromatic",
}
aa_sasa_df = pd.read_csv("./aa_surface_area.csv")
```

```python
data_df = pd.read_csv("./NGS_w_entire_sequence_space.csv")
display_all(data_df)
```

```python
print(data_df.shape)
data_df = data_df.sort_values("sample_name").drop_duplicates("pseudo_prot")
print(data_df.shape)
```

```python
clean_df = data_df[["pseudo_prot", "sample_name", "num_muts", "count", "assign", "predicted"] + all_poss]
clean_df
```

```python
clean_df["num_muts"] = clean_df["pseudo_prot"].progress_map(
    lambda x: len([a for a, p in zip(x, all_poss) if a != p[0]])
)
clean_df[["pseudo_prot", "num_muts", "assign", "predicted"] + all_poss]
```

```python
for pos in tqdm(all_poss):
    wt_cat = AA_CAT[pos[0]]
    wt_sasa = aa_sasa_df.loc[aa_sasa_df["aa"] == pos[0], "sc"].values[0]

    muts_dsasa = {
        aa: wt_sasa - aa_sasa_df.loc[aa_sasa_df["aa"] == aa, "sc"].values[0]
        for aa in clean_df[pos].unique()
    }

    clean_df[f"{pos}_cat_change"] = clean_df[pos].map(lambda x: AA_CAT[x] != wt_cat)
    clean_df[f"{pos}_cat"] = clean_df[pos].map(AA_CAT)
    clean_df[f"{pos}_dSASA"] = clean_df[pos].map(lambda x: muts_dsasa[x])
    if len(clean_df[f"{pos}_cat_change"].unique()) < 2:
        clean_df = clean_df.drop(columns=f"{pos}_cat_change")

helix_poss = ["V61", "V68", "Q69", "S72"]
clean_df["changes_on_helix"] = clean_df.progress_apply(
    lambda r: sum([1 for p in helix_poss if p[0] != r[p][0]]), axis=1
)
on_the_back_poss = ["Q94", "V112", "N121"]
clean_df["changes_on_back"] = clean_df.progress_apply(
    lambda r: sum([1 for p in on_the_back_poss if p[0] != r[p][0]]), axis=1
)
on_the_front_poss = ["Y145", "H148", "T203", "S205", "E222"]
clean_df["changes_on_front"] = clean_df.progress_apply(
    lambda r: sum([1 for p in on_the_front_poss if p[0] != r[p][0]]), axis=1
)
on_hbonds_poss = ["Q94", "H148", "T203", "E222"]
clean_df["changes_on_hbonds"] = clean_df.progress_apply(
    lambda r: sum([1 for p in on_hbonds_poss if p[0] != r[p][0]]), axis=1
)
```

```python
pssm_file = Path("./2wur_pdb_seq.pssm")
pssm_cols = [
    "index",
    "aa",
    "A",
    "R",
    "N",
    "D",
    "C",
    "Q",
    "E",
    "G",
    "H",
    "I",
    "L",
    "K",
    "M",
    "F",
    "P",
    "S",
    "T",
    "W",
    "Y",
    "V",
    "A.1",
    "R.1",
    "N.1",
    "D.1",
    "C.1",
    "Q.1",
    "E.1",
    "G.1",
    "H.1",
    "I.1",
    "L.1",
    "K.1",
    "M.1",
    "F.1",
    "P.1",
    "S.1",
    "T.1",
    "W.1",
    "Y.1",
    "V.1",
    "num1",
    "num2",
]

pssm_df = df = pd.read_csv(pssm_file, sep="\s+", skiprows=3, names=pssm_cols)
correction = -2
for pos in tqdm(all_poss):
    if pos == "T65":
        correction = -4
        continue
    assert (
        pos[0]
        == pssm_df.loc[pssm_df["index"] == int(pos[1:]) + correction, "aa"].values[0]
    )
    wt_pssm = pssm_df.loc[pssm_df["index"] == int(pos[1:]) + correction, pos[0]].values[
        0
    ]
    mut_pssm = {
        aa: wt_pssm
        - pssm_df.loc[pssm_df["index"] == int(pos[1:]) + correction, aa].values[0]
        for aa in clean_df[pos].unique()
    }
    clean_df[f"{pos}_dPSSM"] = clean_df[pos].map(lambda x: mut_pssm[x])
```

```python
clean_df["mean_dSASA"] = clean_df[
    [c for c in clean_df.columns if c.endswith("dSASA")]
].mean(axis=1)
clean_df["mean_dPSSM"] = clean_df[
    [c for c in clean_df.columns if c.endswith("dPSSM")]
].mean(axis=1)
```

```python
comb_wt_seq_space = {
    42: "L",
    44: "L",
    61: "V",
    65: "T",
    68: "V",
    69: "Q",
    72: "S",
    94: "Q",
    108: "T",
    112: "V",
    121: "N",
    145: "Y",
    148: "H",
    150: "V",
    167: "T",
    181: "H",
    185: "N",
    203: "T",
    205: "S",
    220: "L",
    222: "E",
    224: "V",
}
bub_dct = {
    42: [42, 44, 68, 220, 224],
    44: [44, 42, 46, 64, 68, 220, 14, 16],
    46: [46, 44, 64, 68, 220, 16, 123],
    61: [61, 46, 64, 66, 145, 220],
    64: [64, 44, 46, 61, 220, 16, 18, 123],
    68: [68, 42, 44, 112, 224, 14, 119],
    69: [69, 150, 163, 165, 201, 72, 185],
    110: [110, 112, 108, 123],
    112: [112, 68, 110, 119, 185],
    145: [145, 61, 66, 167, 181, 220],
    150: [150, 66, 69, 163, 165, 167, 201, 224],
    163: [163, 69, 150, 165, 181, 201, 185],
    165: [165, 69, 150, 163, 167, 181],
    167: [167, 66, 145, 165, 181],
    181: [181, 145, 150, 165, 167],
    201: [201, 69, 150, 163, 224, 72],
    220: [220, 42, 44, 46, 61, 64, 66, 68, 145],
    224: [224, 42, 68, 69, 201, 72],
    14: [14, 42, 44, 68, 16, 119],
    16: [16, 44, 46, 64, 68, 14, 18, 119, 123],
    18: [18, 46, 64, 16, 108, 123],
    72: [72, 42, 201, 224],
    108: [108, 110, 18, 123],
    119: [119, 68, 112, 14, 16],
    123: [123, 64, 110, 112, 16, 18, 108],
    185: [185, 69, 112, 163],
}
pos_bub_dct = {}
for k, v in bub_dct.items():
    if k not in comb_wt_seq_space.keys():
        continue
    pos_bub_dct[f"{comb_wt_seq_space[k]}{k}"] = [
        f"{comb_wt_seq_space[p]}{p}" for p in v if p in comb_wt_seq_space.keys()
    ]
pos_bub_dct
```

```python
def count_muts_in_bubble(row: pd.Series, poss) -> int:
    count = 0
    for pos in poss:
        if row[pos] != pos[0]:
            count += 1
    return count


for pos, bub in tqdm(pos_bub_dct.items()):
    clean_df[f"bub_{pos}"] = clean_df.apply(
        lambda x: count_muts_in_bubble(x, bub), axis=1
    )
```

```python
clean_df["bub_max"] = clean_df[
    [c for c in clean_df.columns if c.startswith("bub_")]
].max(axis=1)
clean_df["bub_mean"] = clean_df[
    [c for c in clean_df.columns if c.startswith("bub_")]
].mean(axis=1)
```
